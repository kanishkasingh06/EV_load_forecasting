```python
import numpy as np
from scipy.stats import lognorm

def area_under_lognormal(x1, x2, mean, std_dev):
    """
    Calculate the area under the log-normal distribution between x1 and x2.

    Parameters:
        x1 (float): Lower bound.
        x2 (float): Upper bound.
        mean (float): Mean of the log-normal distribution.
        std_dev (float): Standard deviation of the log-normal distribution.

    Returns:
        float: Area under the log-normal distribution between x1 and x2.
    """
    mean_income =mean
    std_income = std_dev

# Calculate parameters for the underlying normal distribution
    mu = np.log(mean_income**2 / np.sqrt(std_income**2 + mean_income**2))
    sigma = np.sqrt(np.log(std_income**2 / mean_income**2 + 1))
    # Calculate the scale parameter using the mean
    scale = np.exp(mean)

    # Calculate the CDF values for x1 and x2
    cdf_x1 = lognorm.cdf(x1, sigma, scale=scale)
    cdf_x2 = lognorm.cdf(x2, sigma, scale=scale)

    # Compute the area between x1 and x2
    area = cdf_x2 - cdf_x1

    return area

# Example usage:
x1 = 0  # lower bound
x2 = 50000  # upper bound
mean = 90000  # mean of the log-normal distribution
std_dev = 24026  # standard deviation of the log-normal distribution

area = area_under_lognormal(x1, x2, mean, std_dev)
print("The area between x1 and x2 is:", area)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

def plot_lognormal_cdf(mean, std_dev):
    """
    Plot the CDF of a log-normal distribution given the mean and standard deviation.

    Parameters:
        mean (float): Mean of the log-normal distribution.
        std_dev (float): Standard deviation of the log-normal distribution.
    """
    # Generate a range of x values
    x = np.linspace(0, 5, 1000)  # Adjust the range and number of points as needed

    # Calculate the scale parameter using the mean
    scale = np.exp(mean)

    # Compute the CDF values
    cdf = lognorm.cdf(x, std_dev, scale=scale)

    # Plot the CDF
    plt.plot(x, cdf, label=f'CDF\nMean: {mean}, Std Dev: {std_dev}')
    plt.title('Log-normal Distribution CDF')
    plt.xlabel('x')
    plt.ylabel('CDF')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
mean = 90000  # mean of the log-normal distribution
std_dev = 24026  # standard deviation of the log-normal distribution

plot_lognormal_cdf(mean, std_dev)
```

```python
import numpy as np
from scipy.stats import lognorm

def proportion_between_incomes(I1, I2, avg_income, std_dev_income):
    """
    Calculate the proportion of people with income between I1 and I2 in a log-normal distribution.
    
    Parameters:
        I1 (float): Lower income bound.
        I2 (float): Upper income bound.
        avg_income (float): Average income (mean of the log-normal distribution).
        std_dev_income (float): Standard deviation of the income (log-normal distribution).
    
    Returns:
        float: Proportion of people with income between I1 and I2.
    """
    # Calculate the mean (mu) and standard deviation (sigma) of the underlying normal distribution
    mu = np.log(avg_income**2 / np.sqrt(std_dev_income**2 + avg_income**2))
    sigma = np.sqrt(np.log(1 + (std_dev_income**2 / avg_income**2)))
    
    # Calculate the CDF values for I1 and I2
    cdf_I1 = lognorm.cdf(I1, sigma, scale=np.exp(mu))
    cdf_I2 = lognorm.cdf(I2, sigma, scale=np.exp(mu))
    
    # Compute the proportion of people with income between I1 and I2
    proportion = cdf_I2 - cdf_I1
    
    return proportion

# Example usage:
I1 = 30000  # Lower income bound
I2 = 70000  # Upper income bound
avg_income = 50000  # Average income
std_dev_income = 15000  # Standard deviation of income

proportion = proportion_between_incomes(I1, I2, avg_income, std_dev_income)
print("The proportion of people with income between I1 and I2 is:", proportion)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

def calculate_lognormal_params(avg_income, std_dev_income):
    """
    Calculate the parameters of the underlying normal distribution from the log-normal distribution.
    
    Parameters:
        avg_income (float): Average income (mean of the log-normal distribution).
        std_dev_income (float): Standard deviation of the income (log-normal distribution).
    
    Returns:
        (float, float): Mean (mu) and standard deviation (sigma) of the underlying normal distribution.
    """
    mu = np.log(avg_income**2 / np.sqrt(std_dev_income**2 + avg_income**2))
    sigma = np.sqrt(np.log(1 + (std_dev_income**2 / avg_income**2)))
    return mu, sigma

def plot_lognormal_distribution(avg_income, std_dev_income):
    """
    Plot the PDF and CDF of a log-normal distribution given the average income and standard deviation.
    
    Parameters:
        avg_income (float): Average income (mean of the log-normal distribution).
        std_dev_income (float): Standard deviation of the income (log-normal distribution).
    """
    # Calculate the mean (mu) and standard deviation (sigma) of the underlying normal distribution
    mu, sigma = calculate_lognormal_params(avg_income, std_dev_income)
    
    # Generate a range of x values
    x = np.linspace(0, avg_income * 3, 1000)
    
    # Compute the PDF and CDF values
    pdf = lognorm.pdf(x, sigma, scale=np.exp(mu))
    cdf = lognorm.cdf(x, sigma, scale=np.exp(mu))
    
    # Plot the PDF
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.plot(x, pdf, label=f'PDF\nMean: {avg_income}, Std Dev: {std_dev_income}')
    plt.title('Log-normal Distribution PDF')
    plt.xlabel('Income')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    
    # Plot the CDF
    plt.subplot(1, 2, 2)
    plt.plot(x, cdf, label=f'CDF\nMean: {avg_income}, Std Dev: {std_dev_income}')
    plt.title('Log-normal Distribution CDF')
    plt.xlabel('Income')
    plt.ylabel('CDF')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

# Example usage:
avg_income = 90000  # Average income
std_dev_income = 25000  # Standard deviation of income

plot_lognormal_distribution(avg_income, std_dev_income)
```
